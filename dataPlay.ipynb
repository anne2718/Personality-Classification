### Make a Copy of Data to work with

```python
usr_cp = users.copy()
lks_cp = likes.copy()
usr_lks_cp = users_likes.copy()
```

```python
usr_cp.reset_index(inplace=True)
lks_cp.reset_index(inplace=True)
# usr_lks_cp.reset_index(inplace=True)
```

```python
usr_cp = usr_cp.rename(columns = {"index": "user_num"})
lks_cp = lks_cp.rename(columns = {"index": "likes_num"})
# usr_lks_cp = usr_lks_cp.rename(columns = {"index": "items"})
```

## Join tables to find the matching Indices

```python
usr_lks_lks = pd.merge(usr_lks_cp, lks_cp, on="likeid")
usr_lks_lks.head()
```

```python
usr_lks_usr = pd.merge(usr_lks_cp, usr_cp, on="userid")
usr_lks_usr.head()
```

```python
likes_vector = usr_lks_lks.loc[:,'likes_num']
users_vector = usr_lks_usr.loc[:,'user_num']
```
